```python
# Group the DataFrame by country, and count the number of respondents who know Python using the 'apply()' method with a lambda function
python_users = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
count_by_country = df['Country'].value_counts()
count_by_country
```

```python
# Create a new DataFrame by concatenating two existing DataFrames using the 'concat()' method
python_df = pd.concat([count_by_country, python_users], axis = 'columns', sort=False)
python_df.rename(columns={'Country': 'RespondentsNum', 'LanguageWorkedWith': 'NumsKnowPython'}, inplace=True)
```

```python
# Calculate the percentage of respondents who know Python and add it as a new column to the DataFrame
python_df['PctKnowsPython'] = (python_df['NumsKnowPython'] / python_df['RespondentsNum']) * 100
```
