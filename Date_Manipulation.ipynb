```python
import pandas as pd
```

```python

df = pd.read_csv('ETH_1h.csv', parse_dates=['Date'])
```

```python
df.drop(['Unix Timestamp'], axis = 1, inplace=True)
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d %I-%p')
```

```python
df.loc[1, 'Date'].day_name()
```

```python
df['Day of Week'] = df['Date'].dt.day_name()
df
```

```python
filt = (df['Date'] < pd.to_datetime('2020-01-01')) & (df['Date'] >= pd.to_datetime('2019-01-01'))
df.loc[filt]
```
